# Cocoa Disease Detection with Resnet34

## Data

```python
PIL.ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
path = Path('data/cocoa/')
```

```python
tfms = get_transforms(flip_vert=True)
```

```python
np.random.seed(352)
data = ImageDataBunch.from_folder(path, valid_pct=0.2, size=224, ds_tfms=tfms).normalize(imagenet_stats)
```

```python
data.show_batch(3, figsize=(15, 11))
```

# Model

```python
learn = cnn_learner(data, models.resnet34, metrics=accuracy)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 1e-2
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-rn34')
```

```python
learn.load('stage-1-rn34');
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-4, lr/5))
```

```python
learn.save('stage-2-rn34-2')
```

```python
interpreter = ClassificationInterpretation.from_learner(learn)
```

```python
interpreter.plot_top_losses(6)
```
